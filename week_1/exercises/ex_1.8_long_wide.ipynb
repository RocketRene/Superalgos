Step 1 
import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import imageio
import os
%matplotlib inline
```

Step 2
Load the file data/gapminder_total_fertility.csv & fertility data into pandas.

```python
life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
pop = pd.read_excel('gapminder_population.xlsx', index_col=0)
fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
```

Step 3
Check the shape of both tables:

```python
life.shape
```

```python
fert.shape
```

Step 4
Take a look at the columns of both tables:

```python
fert.columns
```

```python
life.columns
```

One table has columns as strings, the other has integer numbers. 
To merge the tables, we need to convert both into the same format:

```python
fert.columns = fert.columns.astype(int)
```

Step 5
Now, let’s check the row index of the table:

```python
fert.index
```

The name of the index should be called country. You can change it with:

```python
fert.index.name = 'country'
```

Step 6
Now convert the table into long format. First, we have to move the row index into a column:

```python
fert.reset_index(inplace=True)
```

Second, we use the convenient melt function for the conversion:

```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
```

Step 7
Repeat steps 5-6 for the life expectancy data, so that it’s also in long format.

```python
life.index
```

```python
life.index.name = 'country'
```

```python
life = life.reset_index()
```

```python
life = life.melt(id_vars='country', var_name='year', value_name='life expectancy')
```

Step 8
To make things even more interesting, let’s also introduce population data from the 
file data/gapminder_population.xlsx.
Read the data in with pandas, and get that DataFrame into the same format as 
the other two.

```python
pop.index.name = 'country'
```

```python
pop = pop.reset_index()
```

```python
pop = pop.melt(id_vars='country', var_name='year', value_name='population')
```

Step 9
Now we can start gluing the DataFrames together. This works great with the merge function:

```python
df = fert.merge(pop)
```

```python
df = life.merge(df)
```

```python
df
```

Step 10
With all the data merged into a single object, you can now select specific columns and plot them:

```python
df_subset = df.loc[df['country'].isin(['France', 'Germany', 'Sweden'])]
sns.lineplot(x='year', y='life expectancy', hue='country',
            data=df_subset);
```

You can also produce a scatterplot:

```python
sns.scatterplot(x='year', y='life expectancy', hue='country',
            data=df_subset, alpha=0.6);
```

```python
# testing for step 11
sns.scatterplot(x='fertility_rate', y='life expectancy', hue='country',
            data=df_subset, alpha=0.6);
```

Step 11

To create an animated scatterplot showing the correlation between life expectancy and fertility, create one plot for each year between 1960 and 2015 (before that, the data contains too many gaps).
Write a for loop that goes through each year and produces a scatterplot.
Save each scatterplot to a separate file containing the year in the filename, e.g. lifeexp_1999.png. After saving the plot call plt.close() to remove the plot from the notebook.
Hint Use the function plt.axis((xmin, xmax, ymin, ymax)) to fix the coordinates of the plot.

```python
for i in range(1960,2016):
    sns.scatterplot(x='fertility_rate', y='life expectancy', hue='country',
                    data=df_subset[df_subset['year'] == i], alpha=0.6)
    plt.axis((0,6,25,85))
    plt.savefig(f'lifeexp_{i}')
    plt.close()
```

Step 12

Add another dimension to the visualization by utilizing the population data.
Adjust the plot, so that the size of the symbols represent the population from the gapminder dataset.

```python
for i in range(1960,2016):
    sns.scatterplot(x='fertility_rate', y='life expectancy', hue='country',
                    data=df_subset[df_subset['year'] == i], alpha=0.6, 
                    size='population')
    plt.axis((0,10,0,85))
    plt.savefig(f'lifeexp_part_{i}')
    plt.close()
```

Step 13
To generate a gif file from the individual pictures, 
adjust the following code example and execute it:

```python
images = []

for i in range(1965, 2016):
    filename = 'lifeexp_part_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
```
